2. Attempt to to download the images for all movies available in Thailand.

    c. Download the images on column `Images`. 
  
    d. If the download was succesful, then add the local path to the column `Local images`. Add `None` otherwise.

```python
import requests 
import time
import os
```

```python
#INSERT CODE HERE
# get all movies available in Thailand
df_Thailand = df.loc[(df['Country Availability'].str.contains('Thailand')==True) & (df['Series or Movie']=='Movie')]
print(df_Thailand.shape)

# create a folder Netflix on Google Drive
try:
    os.mkdir("/content/drive/My Drive/38610/Netflix")
except:
    print("Folder already found")

# add an column for image path
df_Thailand["Local Images"] = None
df["Local Images"] = ''
```

```python
df_Thailand.tail()
```

```python
# Loop through all movies available in Thailand to download and save images
start = time.time()

for i in df_Thailand.index:
  image_link = df_Thailand['Image'][i]
  response = requests.get(image_link, stream = True)
  if response.status_code == 200: # if link available and download successful
    save_path = '/content/drive/My Drive/38610/Netflix/' + str(i) +'.png'
    df_Thailand["Local Images"][i] = save_path
    file = open(save_path, "wb")
    file.write(response.content)
    file.close()
    df_Thailand["Local Images"][i] = save_path
    df['Local Images'][i] = save_path
  else:
    df_Thailand["Local Images"][i] = None
    df['Local Images'][i] = None

end = time.time()
print('Time taken: ', end - start)
```

3. Add a column to the table named `IMDb Availability`. If the link `IMDb Link` is unavailable, then populate the record as False. True otherwise.

```python
!pip install validators
```

```python
#INSERT CODE HERE
import validators
import requests 
import time

# Add new column 'IMDb Availability' to df
df["IMDb Availability"] = None

start = time.time()
# check if link is valid and available
for i in df.index:
  # to know the process as it runs too long
  if i % 1000 == 0:
    print(i)

  imdb_link = df['IMDb Link'][i]
  if isinstance(imdb_link, str): # check if record is string
    valid = validators.url(imdb_link)
    if valid == True: # check if link is valid
      response = requests.head(imdb_link)
      if response.status_code == 200 or response.status_code == 308: # check if link is available
        df['IMDb Availability'][i] = True
      else:
        df['IMDb Availability'][i] = False
    else:
      df['IMDb Availability'][i] = False
  else:
    df['IMDb Availability'][i] = False

end = time.time()
print('Time taken: ', end - start)
```

4. Make a histogram using each unique genre in the `Genre` column.

```python
#INSERT CODE HERE
import split

# get unique genre and frequency from 'Genre' column
genre_list = []
count = []
start = time.time()
for i in df.index:
  txt = df['Genre'][i]
  if isinstance(txt, str):
    temp = txt.split(", ")
    for g in temp:
      if (g not in genre_list) and g != '':
        genre_list.append(g)
        count.append(1)
      else:
        idx = genre_list.index(g)
        count[idx] = count[idx] + 1

end = time.time()
print('Time taken: ', end-start)

# add lists of genre to a dataframe
temp_dict = {'Genre':genre_list,'Count':count}
unique_genre = pd.DataFrame(temp_dict)
unique_genre
```
